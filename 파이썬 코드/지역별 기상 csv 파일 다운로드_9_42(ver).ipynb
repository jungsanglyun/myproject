```python
#지역별 자동 다운로드
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from dateutil.relativedelta import relativedelta
from datetime import datetime
from nltk.tokenize import word_tokenize 
import time
import os, shutil

now = datetime.now()

today_str=str(now)
today_day_and_time=[]
today_day_and_time=word_tokenize(today_str)
today_day=today_day_and_time[0].split('-')

year=today_day[0] # 현재년
month=today_day[1]# 현재월
day=today_day[2]# 현재일


before_one_year = now - relativedelta(years=1)

before_today_str=str(before_one_year)
before_day_and_time=[]
before_day_and_time=word_tokenize(before_today_str)
beforey_day=before_day_and_time[0].split('-')

before_year=beforey_day[0]# 저번년
before_month=beforey_day[1]# 저번월
before_day=beforey_day[2]# 저번일



weather_code={'weather_01':'기온','weather_02':'강수량','weather_03':'바람'}
loacation_code={'loacation_01':'서울','loacation_02':'부산','loacation_03':'대구','location_05':'광주','location_06':'대전'}



loacation_plus_xpath={'loacation_01':'14','loacation_02':'17','loacation_03':'19','location_05':'26','location_06':'28'}
loacation_xpath={'loacation_01':'16','loacation_02':'18','loacation_03':'20','location_05':'27', 'location_06':'29'}

filepath = 'C:\\Users\\0\\Downloads' #여러분들의 다운로드 폴더가 있는경로

driver = webdriver.Chrome('./chrome/chromedriver')#여러분들의 크롬드라이버가 있는 경로, 설정해도 안될시 파일.exe까지 작성
driver.implicitly_wait(5)
driver.get('https://data.kma.go.kr/climate/RankState/selectRankStatisticsDivisionList.do?pgmNo=179')
driver.find_element_by_xpath('//*[@id="dataFormCd"]').send_keys('일')
driver.implicitly_wait(5)
driver.find_element_by_xpath('//*[@id="startYear"]').send_keys('2017')#1년 계산해서 시작날짜와 끝날날짜 지정
driver.implicitly_wait(5)
driver.find_element_by_xpath('//*[@id="endYear"]').send_keys('2017')
driver.implicitly_wait(5)

for i in range(2):
    if i == 0:

        for a in loacation_code:
            id1_cnt=1
            id2_cnt=2
            for weather in weather_code: 
                        if id1_cnt==3:
                            id2_cnt=1
                        print(before_year+loacation_code[a]+weather_code[weather]+"데이터 다운로드중")
                        
                        driver.find_element_by_xpath('//*[@id="ureaType"]').send_keys(weather_code[weather])#기상 코드와 기상으로 나누어서 코드와 기상을 map{id:value}로
                        time.sleep(3)
                        driver.find_element_by_xpath('//*[@id="btnStn"]').click()
                        time.sleep(5) 
                        driver.find_element_by_xpath('//*[@id="ztree_'+ loacation_plus_xpath[a]+'_switch"]').click()
                        driver.implicitly_wait(5)
                        driver.find_element_by_xpath('//*[@id="ztree_'+loacation_xpath[a]+'_a"]/label').click()
                        driver.implicitly_wait(5)
                        driver.find_element_by_xpath('//*[@id="sidetreecontrol"]/a').click()
                        time.sleep(3)
                
                        driver.find_element_by_xpath('//*[@id="startYear"]').send_keys(before_year)#이전년도 시작 년
                        driver.implicitly_wait(5)
                        driver.find_element_by_xpath('//*[@id="endYear"]').send_keys(before_year)#이전년도 종료 년
                        driver.implicitly_wait(5)
                
                        driver.find_element_by_xpath('//*[@id="startMonth"]').send_keys(before_month)#이전년도 시작 월
                        driver.implicitly_wait(5)
                        driver.find_element_by_xpath('//*[@id="endMonth"]').send_keys('12')#이전년도 종료월
                        driver.implicitly_wait(5)
                
                        driver.find_element_by_xpath('//*[@id="startDay"]').send_keys('01')#이전년도 종료일
                        driver.implicitly_wait(5)
                        driver.find_element_by_xpath('//*[@id="endDay"]').send_keys('31')#이전년도 종료일
                        driver.implicitly_wait(5)
                
                        driver.find_element_by_xpath('//*[@id="schForm"]/div[2]/button').click()
                        time.sleep(5)
                        
                        
                        
                        driver.find_element_by_xpath('//*[@id="div_0'+str(id1_cnt)+'\"'+']/div['+str(id2_cnt)+']/div/a[1]').click()
                        time.sleep(5)
                        #파일 이름 자동 지정
                        
                    
                        new_filename=before_year+loacation_code[a]+weather_code[weather]+".csv"
                        
                        
                        
                        filename = max([filepath + '\\' + f for f in os.listdir(filepath)], key=os.path.getctime)
                        shutil.move(os.path.join(filepath, filename), new_filename)
                        
                        id1_cnt+=1
                        
                        print(before_year+loacation_code[a]+weather_code[weather]+"데이터 다운로드 완료")
    else:
        for a in loacation_code:
            id1_cnt=1
            id2_cnt=2
            for weather in weather_code: 
                        if id1_cnt==3:
                            id2_cnt=1
                        print(year+loacation_code[a]+weather_code[weather]+"데이터 다운로드중")
                        
                        driver.find_element_by_xpath('//*[@id="ureaType"]').send_keys(weather_code[weather])#기상 코드와 기상으로 나누어서 코드와 기상을 map{id:value}로
                        time.sleep(3)
                        driver.find_element_by_xpath('//*[@id="btnStn"]').click()
                        time.sleep(5) 
                        driver.find_element_by_xpath('//*[@id="ztree_'+ loacation_plus_xpath[a]+'_switch"]').click()
                        driver.implicitly_wait(5)
                        driver.find_element_by_xpath('//*[@id="ztree_'+loacation_xpath[a]+'_a"]/label').click()
                        driver.implicitly_wait(5)
                        driver.find_element_by_xpath('//*[@id="sidetreecontrol"]/a').click()
                        time.sleep(3)
                
                        driver.find_element_by_xpath('//*[@id="startYear"]').send_keys(year)#이전년도 시작 년
                        driver.implicitly_wait(5)
                        driver.find_element_by_xpath('//*[@id="endYear"]').send_keys(year)#이전년도 종료 년
                        driver.implicitly_wait(5)
                
                        driver.find_element_by_xpath('//*[@id="startMonth"]').send_keys('01')#이전년도 시작 월
                        driver.implicitly_wait(5)
                        driver.find_element_by_xpath('//*[@id="endMonth"]').send_keys(month)#이전년도 종료월
                        driver.implicitly_wait(5)
                
                        driver.find_element_by_xpath('//*[@id="startDay"]').send_keys('01')#이전년도 종료일
                        driver.implicitly_wait(5)
                        driver.find_element_by_xpath('//*[@id="endDay"]').send_keys('31')#이전년도 종료일
                        driver.implicitly_wait(5)
                
                        driver.find_element_by_xpath('//*[@id="schForm"]/div[2]/button').click()
                        time.sleep(5)
                        
                        
                        
                        driver.find_element_by_xpath('//*[@id="div_0'+str(id1_cnt)+'\"'+']/div['+str(id2_cnt)+']/div/a[1]').click()
                        time.sleep(5)
                        #파일 이름 자동 지정
                        
                        new_filename=year+loacation_code[a]+weather_code[weather]+".csv"
                        
                        
                        filename = max([filepath + '\\' + f for f in os.listdir(filepath)], key=os.path.getctime)
                        shutil.move(os.path.join(filepath, filename), new_filename)
                        
                        id1_cnt+=1
                        print(year+loacation_code[a]+weather_code[weather]+"데이터 다운로드 완료")
                        
print("데이터 다운로드 완료")
```
